## Introduction
Broadly speaking, machine learning is a sub-field of artificial intelligence and statistics that deals with the automatic extraction of patterns (i.e. learning) from large amounts of data. We may broadly categorize the types of learning as either __supervised__ or __unsupervised__ depending on whether the data we are given is labelled or not (note: there are also other types of learning such as reinforcement learning that deals with learning from past mistakes, or semi-supervised learning, where only a subset of the data is labelled).

![Different types of machine learning](../images/ml-types.png)

In this tutorial, we will only be concerned with supervised learning tasks, so our data will be *labelled*, by which we mean that it consists of a pair $(X_n, y_n)_{n=1}^N$ of __inputs__ $X_n$ and __labels__ $y_n$. For example, the inputs here may be images of cats and dogs, and the labels are the tags 'cat' or 'dog' corresponding to the images.

Now, the primary task of supervised learning is to find a suitable mapping $f : X \mapsto Y$ that best represents the relationship between the inputs and labels of our dataset, which, in statistics lingo, is called **fitting the data**. Depending on whether the labels $y_n$ take values in a discrete or continuous set, the set of techniques used for fitting the data are slightly different. The tasks also have different names: the former is called __classification__ and the latter is called **regression**.

This tutorial will cover both classification and regression tasks, and will be organised as follows:
1. Introduction to machine learning for classification tasks
2. Introduction to machine learning for regression tasks
3. Handling real world datasets


## Part 1: Classification task with the Lorenz 63 system

In the first part of the tutorial, we are going to get familiar with the basic tools used for statistical classification and apply it to simulated data from the well-known Lorenz 63 model.

First, we import some basic python modules that will be used.

```python
import numpy as np
import matplotlib.pyplot as plt
import os, sys
currentdir = os.path.dirname(os.path.realpath('__file__'))
parentdir = os.path.dirname(currentdir)
sys.path.append(parentdir)
from datasets.lorenz import Lorenz63
```

The Lorenz 63 model (often simplified as L63) is given by the following three-dimensional set of nonlinear ODEs, introduced by the meteorologist Edward Lorenz in 1963 as a toy model for atmospheric convection:

\begin{align*}
\frac{dx}{dt} &= \sigma (y - x), \\
\frac{dy}{dt} &= x(\rho - z) - y, \\
\frac{dz}{dt} &= xy - \beta z.
\end{align*}

This innocent looking model is famous for kickstarting the field of _chaos theory_ that has since become important in the physical sciences and especially in meteorology, demonstrating that it is practically impossible to forecast the weather far into the future.

This model has three parameters $(\sigma, \rho, \beta)$, which are usually set at $\sigma = 10, \rho = 28$ and $\beta = 8/3$, at which the system exhibits chaotic behaviour.

Below, we integrate and visualise the L63 system with the parameters set at the afforementioned values.

```python
# Generate and plot the Lorenz attractor
tmax = 100
num_samples = 10000
dt = tmax/num_samples # 0.01
X0 = np.array([0, 1, 1.05]) # Initial condition

L63 = Lorenz63(tmax, X0, num_samples)
L63.plot()
```

The famous butterfly! A fitting visual considering that chaos also has the nickname 'butterfly effect'. This butterfly is called the *Lorenz attractor*.

Now let's say that for a point $X_0 = (x_0,y_0,z_0)$ on the Lorenz attractor, we want to predict where it will land after time $t=1$. Further, let's assume that we have no knowledge of the L63 equations and only have noisy measurements of its (x,y,z)-coordinates. Is this even possible?

Due to the chaotic nature of the L63 model, predicting the exact position is going to be virtually impossible so we will simplify the problem a bit to make it a little more tractable:

__Challenge:__ Given any point $X_0$ on the Lorenz attractor, determine which wing of the butterfly (left or right) it will end up in after time $t=1$.

Here, we define the right/left butterfly wing of the Lorenz attractor according to whether the x-component of $X$ is positive or negative respectively. You should realise that this is a _classification task_ since the labels 'left butterfly wing' and 'right butterfly wing' form a discrete set.

To proceed, let us first construct the following function that takes in a sequence $X_{n} = (x_{n}, y_{n}, z_{n}), \, n = 1, \ldots, N$ of the L63 trajectory and generates a sequence of labels (0 or 1) depending on whether the future point $X_{n + \Delta n}$ is on the left or right butterfly wing respectively, for fixed $\Delta n \in \mathbb{N}$.

```python
def which_wing(X: np.array, lag: int) -> np.array:
    """ Returns array of 0s and 1s corresponding to whether a point on the attractor lands on the
        left or right wing respectively after n = lag timesteps
        Args:
            X: array of shape (num_samples, 3)
        
        Output: array of shape (num_samples - lag, )
    """
    x, y, z = X.transpose()
    return (x[lag:] > 0).astype(int)
```

We use this to generate the sequence of labels for our task.

```python
tfinal = 1.
tsteps = int(tfinal/dt)

X = L63.trajectory
N = L63.num_samples

# Get sequence of 0/1s depending on whether each point X ends up on the left/right wing after time t=1
lag = tsteps
labels = which_wing(X,lag=lag)
```

Next, we visualise our task by color coding the points on the attractor: orange if the point ends up on the left butterfly wing after time $t=1$ and blue if it ends up on the right butterfly wing.

```python
# Get indices corresponding to left/right wing after time t=1
left_indices = np.argwhere(labels == 0)
right_indices = np.argwhere(labels == 1)

# Extract subset of X that lies on the left/right wing after time t=1
eventually_left_X = np.array([X[i] for i in left_indices]).squeeze()
eventually_right_X = np.array([X[i] for i in right_indices]).squeeze()

# Plot
fig = plt.figure(figsize=(6, 4))
ax = plt.gca()
left_x, left_z = eventually_left_X[:,0], eventually_left_X[:,2]
right_x, right_z = eventually_right_X[:,0], eventually_right_X[:,2]
plt.plot(left_x, left_z, 'o', color='tab:orange', ms=2, alpha=0.6, label=f'left at t={int(tfinal)}')
plt.plot(right_x, right_z, 'o', color='tab:blue', ms=2, alpha=0.6, label=f'right at t={int(tfinal)}')
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('z', fontsize=15)
ax.set_title(f't = 0', fontsize=15)
ax.legend()
plt.show()
```

We can see from this that our objective is to learn a function $f$ that best classifies points on the attractor into the orange or blue regions above.

Below, we construct our labelled dataset $(X, y)$ where the inputs $X$ are noisy observations (assumed to be i.i.d. Gaussian) of points on the Lorenz attractor and the labels $y$ are the corresponding 0/1 class.

```python
# noise level
sigma = 0.1

# Dataset for our task
X = X[:N-lag] + sigma * np.random.randn(N-lag, 3)
y = labels
```

In the following sections, we explore three common algorithms that can be used to fit the data $(X, y)$, namely, logistic regression, decision trees classification and neural network classification.

## 1. Logistic Regression

Despite the name logistic 'regression', this is really an algorithm used for classification. The main idea behind logistic regression is to model the probability of an input $X \in \mathbb{R}^m$ belonging to a certain class $y$. For binary classification (i.e. there are only two classes $0$ or $1$), the probability of $X$ belonging to the class $1$ is computed using the so-called __sigmoid function__ $\sigma$ (see figure below), which squashes real numbers to take values between $0$ and $1$.

![Sigmoid graph](https://www.googleapis.com/download/storage/v1/b/kaggle-forum-message-attachments/o/inbox%2F5152157%2F366ce717ad74dfe967a940d9a2161f9b%2Fmain-qimg-6b67bea3311c3429bfb34b6b1737fe0c.jpg?generation=1605711218576073&alt=media)

We usually assume a linear relationship between the inputs and _logits_ (raw outputs before applying the sigmoid function), giving us the model

$$f(X; {\boldsymbol \beta}) = \sigma(\underbrace{\beta_0 + \beta_1 X_1 + \ldots + \beta_m X_m}_{\text{logit}}) \in (0,1),$$

where $\beta_0, \beta_1, \ldots, \beta_m$ are real-valued parameters that are to be determined by minmizing the **cross-entropy loss**:

$$L(y, f(X; {\boldsymbol \beta})) = - \left[y \log f(X; {\boldsymbol \beta}) + (1 - y) \log(1 - f(X; {\boldsymbol \beta}))\right],$$

which is a measure of discrepancy between two probability distributions.

In the case of multi-class classification, its multi-dimensional generalisation called __softmax function__ is used for normalisation.

To fit our data using logistic regression, we will work with python's excellent scikit-learn library.

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, labels, test_size=0.2)
```

```python
model_lr = LogisticRegression()
model_lr.fit(train_X, train_y)
```

```python
def plot_predictions(f, X, y, title):
    ypred = f(X)
    fig = plt.figure(figsize=(20, 4))
    coord = {0: 'x', 1: 'y', 2: 'z'}
    for i in range(3):
        ax = plt.subplot(1, 3, i+1)
        plt.plot(X[:250,i], y[:250], 'o', label='truth')
        plt.plot(X[:250,i], ypred[:250], 'x', alpha=0.8, label='prediction')
        ax.set_xlabel(coord[i], fontsize=15)
        if i==0:
            ax.set_ylabel('class', fontsize=15)
    ax.legend(fontsize=12)
    plt.suptitle(title, fontsize=20)
    plt.show()

plot_predictions(model_lr.predict, test_X, test_y, 'Predictions from Logistic Regression')
```

```python
accuracy_lr = model_lr.score(test_X, test_y)
print("Accuracy of logistic regression: {:.4f}".format(accuracy_lr))
```

### Extending the feature vector

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(2, interaction_only=True, include_bias=False)
transform = poly.fit_transform

model_lr_plus = LogisticRegression()
train_X_extended = transform(train_X)

model_lr_plus.fit(train_X_extended, train_y)
```

```python
test_X_extended = transform(test_X)
accuracy_lr_ext = model_lr_plus.score(test_X_extended, test_y)
plot_predictions(model_lr_plus.predict, test_X_extended, test_y, 'Predictions from Logistic Regression+')
print("Accuracy of logistic regression with extra quadratic features: {:.4f}".format(accuracy_lr_ext))
```

Not bad!

## 2. Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

### 2.1. Selecting a decision tree

```python
model_dt = DecisionTreeClassifier()
model_dt.fit(train_X, train_y)
```

```python
fig = plt.figure(figsize = (20, 10))
sklearn.tree.plot_tree(model_dt,
                       max_depth = 2,
                       feature_names = ['x', 'y', 'z'],
                       class_names = ['left', 'right'],
                       filled = True)
plt.show()
```

```python
accuracy_dt = model_dt.score(test_X, test_y)
plot_predictions(model_dt.predict, test_X, test_y, 'Predictions from Decision Tree')
print("Accuracy of decision tree: {:.4f}".format(accuracy_dt))
```

### 2.2. Random forests

```python
model_rf = RandomForestClassifier(max_depth=15)
model_rf.fit(train_X, train_y)
```

```python
fig = plt.figure(figsize = (20, 10))
for i in range(4):
    ax = plt.subplot(2, 2, i+1)
    sklearn.tree.plot_tree(model_rf.estimators_[i],
                           max_depth = 1,
                           feature_names = ['x', 'y', 'z'],
                           class_names = ['left', 'right'],
                           filled = True)
    ax.set_title(f'tree {i+1}', fontsize=12)
plt.show()
```

```python
plot_predictions(model_rf.predict, test_X, test_y, 'Predictions from Random Forest')
accuracy_rf = model_rf.score(test_X, test_y)
print("Accuracy of random forest: {:.4f}".format(accuracy_rf))
```

## 3. Neural Networks

```python
from tensorflow import keras
from tensorflow.keras import Sequential, Input
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers.experimental.preprocessing import Normalization
```

### 3.1. Data pre-processing

First, we introduce a normalizing layer with respect to the training data. See [this article](https://machinelearningmastery.com/how-to-improve-neural-network-stability-and-modeling-performance-with-data-scaling/) for more details on why it is a good idea to normalize data.

```python
def box_plot(X: np.array, title: str):
    # X must have size (n_samples, 3)
    fig = plt.figure(figsize=(6, 4))
    ax = fig.gca()
    plt.boxplot(X, labels=('x', 'y', 'z'), showfliers=False)
    ax.tick_params(axis='both', labelsize=12)
    plt.title(title, fontsize=14)
    plt.show()

box_plot(train_X, 'Distribution of training data before normalization')
```

```python
Normalize = Normalization()
Normalize.adapt(train_X)
print(f'mean of training data: {Normalize.mean.numpy()}')
print(f'variance of training data: {Normalize.variance.numpy()}')
```

```python
train_X_normalized = Normalize(train_X)
box_plot(train_X_normalized.numpy(), 'Distribution of training data after normalization')
```

Much better.

__Note:__ The test dataset will also be normalized using the mean and variance of the training dataset.

### 3.2. Setting up a neural network with keras

```python
model_nn = Sequential([
    Input(shape=(3,)),
    Normalize,
    Dense(100, activation='relu'),
    Dense(100, activation='relu'),
    Dense(100, activation='relu'),
    Dense(100, activation='relu'),
    Dense(100, activation='relu'),
    Dense(1, activation='sigmoid')])

loss_fn = keras.losses.BinaryCrossentropy()
opt = keras.optimizers.Adam(learning_rate=0.001)

model_nn.compile(loss=loss_fn,
                 optimizer='adam',
                 metrics=['accuracy'])

model_nn.summary()
```

### 3.3. Training and testing

```python
model_nn.fit(train_X, train_y, epochs=10)
```

```python
class to_binary():
    def __init__(self, pred_model, cutoff=0.5):
        self.model = pred_model
        self.cutoff = cutoff
    def predict(self, X):
        ypred = self.model(X)
        ypred_binary = np.array([0 if p < self.cutoff else 1 for p in ypred])
        return ypred_binary

model_nn_binary = to_binary(model_nn)
plot_predictions(model_nn_binary.predict, test_X, test_y, 'Predictions from Neural Network')
```

```python
loss_nn, accuracy_nn = model_nn.evaluate(test_X, test_y)
```

__Tip:__ Experiment with different numbers of hidden layers, units, number of epochs, etc in the neural network and see how it affects the test accuracy. Also observe what happens when you remove the normalization layer from the network. You should see that the performance becomes slightly worse!
