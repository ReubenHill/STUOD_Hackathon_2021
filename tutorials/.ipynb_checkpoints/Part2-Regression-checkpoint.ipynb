```python
import numpy as np
import matplotlib.pyplot as plt
import os, sys
currentdir = os.path.dirname(os.path.realpath('__file__'))
parentdir = os.path.dirname(currentdir)
sys.path.append(parentdir)
from datasets.lorenz import Lorenz96
import time
```

```python
# Generate and plot Lorenz 63 model
tmax, dt = 6, 1e-3
num_samples = int(round(tmax/dt)) #6000
K, J = 36, 10
filename = parentdir+'/datasets/'

# Load pre-generated initial conditions
X0 = np.load(parentdir+'/datasets/L96_initial_conditions/X0.npy')
Y0 = np.load(parentdir+'/datasets/L96_initial_conditions/Y0.npy')
```

```python
# Solve L96 system
L96 = Lorenz96(tmax, X0, Y0, num_samples)
```

```python
L96.plot()
```

```python
h, c, b, F = L96.params
B = -1*h*c*L96.trajectory['Y'].reshape(-1,K,J).mean(-1)
```

```python
plt.scatter(np.ravel(L96.trajectory['X'])[::100], np.ravel(B)[::100], alpha=0.5)
```

```python
X = L96.trajectory['X']
m = np.stack((B[:,0], X[:,0], X[:,1], X[:,2], X[:,3], X[:,4], X[:,5], X[:,6], X[:,7], X[:,8], X[:,9], X[:,10]))
cov = np.corrcoef(m)
```

```python
cov[0]
```

```python
plt.imshow(cov, cmap='RdBu', vmin = -1, vmax = 1)
cb = plt.colorbar()
plt.show()
```

```python
X = L96.trajectory['X']
m = np.stack((B[:,5], X[:,0], X[:,1], X[:,2], X[:,3], X[:,4], X[:,5], X[:,6], X[:,7], X[:,8], X[:,9], X[:,10]))
cov = np.corrcoef(m)

plt.imshow(cov, cmap='RdBu', vmin = -1, vmax = 1)
cb = plt.colorbar()
plt.show()
```
